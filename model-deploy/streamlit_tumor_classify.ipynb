```python

%%writefile app.py
import streamlit as st
import pandas as pd
from PIL import Image
import streamlit as st


import streamlit as st
import pandas as pd
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
from tensorflow.keras import preprocessing
from tensorflow.keras.models import load_model
from tensorflow.keras.activations import softmax
import os
import h5py


def main_page():
    st.sidebar.markdown("###### DETEKSI TANDA AWAL PENYAKIT ALZHEIMER")
    with st.container():
        st.markdown('---')
        st.markdown\
             ("<h1 style='text-align: center; color: #728FCE;'>Deteksi Tanda Awal Penyakit Alzheimer</h1>", unsafe_allow_html=True)
        st.markdown("<h3 style='text-align: center; color: #728FCE;'>(by Silvia Rahma)</h1>", unsafe_allow_html=True)
        st.markdown("<h1 style='text-align: center; color: #728FCE;'>ü¶†‚ö†Ô∏èüñ•Ô∏è</h1>", unsafe_allow_html=True)
        st.markdown('---')

        st.markdown("<h3 style='color: #728FCE;'>Klasifikasi pasien dilakukan berdasarkan metadata yang disediakan dan nilai Clinical Dementia Rating (CDR)</h3>", unsafe_allow_html=True)
        st.markdown("<h3 style='color: #728FCE;'>Terdiri atas 4 kelas tanda-tanda awal penyakit Alzheimer (CDR)</h3>", unsafe_allow_html=True)
        st.markdown('- Dementia'
            '\n- Very mild Dementia'
            '\n- mild Dementia'
            '\n- Non Demented')
        st.markdown("<h3 style='color: #728FCE;'>Fitur</h3>", unsafe_allow_html=True)
        st.markdown('- Deteksi'
            '\n- Definisi'
            '\n- Gejala'
            '\n- Pencegahan'
            '\n- Penanganan')

# alzheimer
def alzheimer():
    st.sidebar.markdown('- Deteksi Tanda Awal Penyakit Alzheimer'
                    '\n- Penjelasan Tanda Awal Penyakit Alzheimer')
    st.markdown('---')
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>Penyakit Alzheimer</h1>", unsafe_allow_html=True)
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>ü¶†</h1>", unsafe_allow_html=True)
    st.markdown('---')

    # Deteksi penyakit
    st.markdown("<h2 style='color: #728FCE;'>Upload Gambar Irisan Otak 2D</h1>", unsafe_allow_html=True)

        #fungsi untuk prediksi kelas
    def predict_class_cabai(image):
        classifier_model = tf.keras.models.load_model(r'/content/drive/MyDrive/portofolio/Streamlit tumor/Hasil/densenet121_alzheimer.hdf5')
        shape = ((416, 416, 3))
        model = tf.keras.Sequential([hub.KerasLayer(classifier_model, input_shape=shape)])

        test_image = image.resize((416, 416))
        test_image = preprocessing.image.img_to_array(test_image)
        #test_image = test_image/225.0
        test_image = np.expand_dims(test_image, axis = 0)
        class_names = ['Mild Dementia', 'Moderate Dementia', 'Non Demented', 'Very mild Dementia']
        prediction = model.predict(test_image)
        score = tf.nn.softmax(prediction[0])
        score = score.numpy()
        image_class =  class_names[np.argmax(score)]
        return image_class

 # fungsi untuk import atau upload gambar
    file_uploaded = st.file_uploader("Pilih File", type = ['jpg', 'png', 'jpeg'])
    if file_uploaded is not None:
        image = Image.open(file_uploaded)
        figure = plt.figure(figsize=(4, 3))
        plt.imshow(image)
        plt.axis('off')
        predict_class = predict_class_cabai(image)
        hasil = "Prediksi penyakit cabai : {}".format(predict_class)
        st.write(hasil)
        st.pyplot(figure)
        # Penjelasan Penyakit
        st.markdown("<h2 style='color: #728FCE;'>Penjelasan Penyakit</h1>", unsafe_allow_html=True)
        if(predict_class=='Mild Dementia'):
            st.write('#### Definisi :')
            st.write(" Mild dementia Alzheimer adalah tahap awal dari Alzheimer, di mana penderitanya mengalami gangguan kognitif dan memori yang memengaruhi kemampuan mereka untuk berfungsi secara normal dalam kehidupan sehari-hari. Gejala pada tahap ini mungkin tampak ringan dan sering kali tidak terlalu mencolok, tetapi tetap mempengaruhi kualitas hidup individu.")
            st.write('#### Gejala :')
            st.write(" - Gangguan memori jangka pendek: Orang dengan mild dementia Alzheimer mungkin kesulitan mengingat peristiwa baru-baru ini atau informasi yang baru saja diberikan kepada mereka."
                   "\n - Kesulitan dalam merencanakan dan mengorganisir: Penderitanya mungkin mengalami kesulitan dalam merencanakan atau menyelesaikan tugas-tugas sehari-hari yang melibatkan langkah-langkah yang kompleks"
                   "\n - Masalah berbicara dan menulis: Kesulitan berbicara, mengikuti percakapan, atau menulis dapat menjadi gejala di tahap ini"
                   "\n - Bingung tentang waktu dan tempat: Penderita mungkin kebingungan tentang hari, tanggal, atau tempat tinggal mereka saat ini."
                   "\n - Kesulitan dalam mengenali wajah atau objek: Penderita dapat mengalami kesulitan mengenali wajah orang-orang terdekat atau benda-benda yang biasa mereka gunakan.")
            st.write('#### Pencegahan:')
            st.write(" - Menerapkan pola makan sehat, seperti pola Mediterania."
                     "\n - Melakukan aktivitas fisik secara rutin."
                     "\n - Berpikir aktif, misalnya dengan membaca, menulis, atau bermain puzzle."
                     "\n - Menjaga tekanan darah, kolesterol, dan gula darah dalam kisaran yang sehat"
                     "\n - Menghindari merokok dan membatasi konsumsi alkohol.")

            st.write('#### Penanganan :')
            st.write(" Mild dementia Alzheimer tidak dapat disembuhkan, tetapi perawatan dan manajemen yang tepat dapat membantu mengatasi gejala dan meningkatkan kualitas hidup penderita. Beberapa strategi penanganan meliput : "
                    "\n - Pengobatan: Dokter dapat meresepkan obat-obatan untuk membantu mengurangi gejala dan memperlambat perkembangan penyakit."
                    "\n - Terapi: Terapi kognitif dan terapi perilaku dapat membantu meningkatkan kemampuan kognitif dan mengatasi perubahan perilaku."
                    "\n - Menggunakan alat bantu: Alat bantu seperti kalender, jam alarm, atau catatan tertulis dapat membantu meningkatkan kemampuan untuk mengingat dan mengorganisir."
                    "\n - Dukungan sosial: Dukungan dari keluarga dan teman-teman sangat penting untuk membantu penderita menghadapi tantangan sehari-hari.")

        elif(predict_class=='Moderate Dementia'):
            st.write('#### Definisi :')
            st.write(" Moderate dementia Alzheimer adalah tahap lanjutan dari Alzheimer di mana penderita mengalami gangguan kognitif dan memori yang semakin parah, menyebabkan kesulitan dalam berpikir, berbicara, berorientasi, dan melakukan tugas-tugas sehari-hari.")
            st.write('#### Gejala :')
            st.write(" - Gangguan memori yang lebih parah: Penderita akan kesulitan mengingat informasi penting tentang diri mereka sendiri dan orang-orang di sekitarnya, termasuk anggota keluarga terdekat."
                   "\n - Kesulitan berbicara dan berkomunikasi: Berbicara menjadi lebih sulit, dan mereka mungkin seringkali kebingungan dalam mencari kata atau mengikuti percakapan."
                   "\n - Kesulitan berorientasi: Penderita dapat menjadi bingung tentang waktu, tempat, dan bahkan identitas diri mereka sendiri"
                   "\n - Kesulitan dalam melakukan tugas sehari-hari: Mengerjakan tugas-tugas sederhana seperti berpakaian, mandi, atau makan menjadi semakin sulit."
                   "\n - Perubahan perilaku dan emosi: Penderita mungkin menunjukkan perubahan emosi yang drastis, seperti kebingungan, kecemasan, marah, atau depresi."
                   "\n - Gangguan tidur: Penderita dapat mengalami gangguan tidur, seperti kesulitan tidur atau tidur terlalu banyak.")
            st.write('#### Pencegahan:')
            st.write(" Belum ada cara untuk mencegah atau menyembuhkan dementia Alzheimer sepenuhnya. Namun, beberapa langkah pencegahan kesehatan dapat membantu mengurangi risiko mengembangkan demensia dan menjaga kesehatan otak:"
                     "\n - Pola makan sehat: Menerapkan pola makan sehat, seperti pola Mediterania yang kaya akan buah, sayuran, kacang-kacangan, biji-bijian, dan ikan, dapat membantu mendukung kesehatan otak."
                     "\n - Aktivitas fisik: Melakukan olahraga dan aktivitas fisik secara rutin dapat membantu meningkatkan aliran darah ke otak dan mendukung kesehatan kognitif."
                     "\n - Aktivitas mental: Terus menantang otak dengan membaca, menulis, bermain puzzle, atau berpartisipasi dalam kegiatan kognitif lainnya dapat membantu menjaga kognisi yang sehat."
                     "\n - Mengelola kondisi kesehatan lainnya: Menjaga tekanan darah, kolesterol, dan diabetes dalam kisaran yang sehat juga dapat membantu mengurangi risiko demensia.")

            st.write('#### Penanganan :')
            st.write(" Penanganan moderate dementia Alzheimer bertujuan untuk meningkatkan kualitas hidup penderita, mengatasi gejala, dan memberikan perawatan yang memadai. Beberapa strategi penanganan meliputi : "
                    "\n - Perawatan medis: Pengobatan dengan obat-obatan tertentu dapat membantu mengurangi beberapa gejala dan memperlambat perkembangan penyakit."
                    "\n - Terapi: Terapi kognitif dan terapi perilaku terus dilanjutkan untuk membantu penderita tetap berfungsi sebanyak mungkin dan mengatasi perubahan perilaku."
                    "\n - Dukungan lingkungan: Menciptakan lingkungan yang aman dan mendukung di rumah atau fasilitas perawatan jangka panjang penting untuk keamanan dan kenyamanan penderita."
                    "\n - Perencanaan perawatan jangka panjang: Pada tahap ini, seringkali diperlukan perencanaan perawatan jangka panjang untuk memastikan penderita mendapatkan perawatan yang sesuai dan disesuaikan dengan kebutuhannya."
                    "\n - Dukungan sosial: Dukungan keluarga, teman, atau anggota komunitas dapat membantu penderita dan keluarga menghadapi tantangan fisik dan emosional yang terkait dengan penyakit ini.")



        elif(predict_class=='Non Demented'):
            st.write('#### Definisi :')
            st.write(" Pada tahap non demented Alzheimer, penderita mungkin belum menunjukkan gejala kognitif yang jelas seperti gangguan memori atau kesulitan berbicara.")
            st.write('#### Gejala :')
            st.write(" - Keluhan ringan mengenai gangguan memori: Mungkin ada keluhan tentang kesulitan mengingat sesuatu, tetapi ini belum mencapai tingkat yang signifikan sehingga mempengaruhi aktivitas sehari-hari."
                   "\n - Gangguan kognitif ringan: Penderita mungkin mengalami kesulitan dalam menyelesaikan tugas-tugas yang membutuhkan kognisi tinggi, seperti mengatur rencana atau melakukan tugas-tugas yang memerlukan pemecahan masalah.")

            st.write('#### Pencegahan:')
            st.write(" Beberapa langkah pencegahan kesehatan dapat membantu mengurangi risiko perkembangan penyakit ini atau memperlambat kemajuannya:"
                     "\n - Pola makan sehat: Konsumsi makanan yang kaya akan antioksidan, seperti buah-buahan dan sayuran, serta makanan yang mengandung asam lemak omega-3, seperti ikan berlemak, dapat mendukung kesehatan otak."
                     "\n - Aktivitas fisik: Olahraga dan aktivitas fisik secara teratur dapat membantu meningkatkan aliran darah ke otak dan mendukung kesehatan kognitif."
                     "\n - Aktivitas mental: Stimulasi otak yang teratur melalui kegiatan mental seperti membaca, menulis, belajar, atau bermain puzzle dapat membantu menjaga kognisi yang sehat."
                     "\n - Mengelola kondisi kesehatan lainnya: Menjaga tekanan darah, kolesterol, dan diabetes dalam kisaran yang sehat juga dapat membantu mengurangi risiko Alzheimer.")

            st.write('#### Penanganan :')
            st.write(" - Pemantauan dan evaluasi rutin: Pemantauan kesehatan dan fungsi kognitif secara teratur dapat membantu mendeteksi perubahan lebih awal. "
                    "\n - Penyuluhan dan dukungan: Memberikan informasi dan dukungan kepada penderita dan keluarga tentang perubahan yang mungkin terjadi di tahap awal penyakit dapat membantu mengatasi kecemasan dan ketidakpastian."
                    "\n - Penyelidikan lebih lanjut: Beberapa studi dan penelitian sedang berlangsung untuk memahami lebih lanjut tentang penyakit Alzheimer pada tahap awal, termasuk upaya untuk mengembangkan tes dan intervensi yang lebih efektif.")

        elif(predict_class=='Very mild Dementia'):
            st.write('#### Definisi :')
            st.write(" Very mild dementia Alzheimer adalah tahap awal dari penyakit Alzheimer, di mana penderita mengalami gangguan kognitif dan memori yang masih sangat ringan, tetapi sudah mulai memengaruhi fungsi sehari-hari mereka")
            st.write('#### Gejala :')
            st.write(" - Gangguan memori ringan: Penderita mungkin mengalami kesulitan mengingat kata-kata, nama orang-orang terdekat, atau peristiwa yang baru saja terjadi."
                   "\n - Kesulitan dalam merencanakan dan mengorganisir: Penderita mungkin membutuhkan lebih banyak waktu atau bantuan untuk merencanakan atau menyelesaikan tugas-tugas yang kompleks."
                   "\n - Kehilangan atau kesulitan menemukan benda-benda yang sering digunakan: Penderita mungkin seringkali meletakkan benda-benda mereka di tempat yang salah atau tidak ingat di mana mereka meletakkannya."
                   "\n - Kesulitan dalam mengikuti arah atau petunjuk: Penderita dapat mengalami kesulitan dalam mengikuti petunjuk arah atau dalam perjalanan yang biasanya mereka lakukan dengan mudah."
                   "\n -Kesulitan dalam mengikuti arah atau petunjuk: Penderita dapat mengalami kesulitan dalam mengikuti petunjuk arah atau dalam perjalanan yang biasanya mereka lakukan dengan mudah. ")

            st.write('#### Pencegahan:')
            st.write(" - Pola makan sehat: Menerapkan pola makan yang sehat dan bergizi, seperti pola Mediterania yang kaya akan buah-buahan, sayuran, biji-bijian, dan lemak sehat, dapat mendukung kesehatan otak."
                     "\n - Aktivitas mental: Tetap aktif secara mental dengan membaca, menulis, memecahkan teka-teki, atau bermain permainan otak dapat membantu menjaga kognisi yang sehat."
                     "\n - Aktivitas fisik: Melakukan olahraga dan aktivitas fisik secara teratur dapat membantu meningkatkan aliran darah ke otak dan mendukung kesehatan kognitif."
                     "\n - Pengelolaan stres: Mengelola stres dengan cara yang sehat, seperti meditasi, yoga, atau aktivitas relaksasi lainnya, dapat membantu melindungi kesehatan otak.")

            st.write('#### Penanganan :')
            st.write(" Pada tahap very mild dementia Alzheimer, penanganan bertujuan untuk mendukung fungsi kognitif yang ada dan memberikan perawatan yang tepat bagi penderita. Beberapa strategi penanganan meliputi: "
                    "\n - Pemantauan dan evaluasi teratur: Pemantauan kesehatan dan fungsi kognitif secara teratur dapat membantu mendeteksi perubahan lebih awal dan menentukan perencanaan perawatan yang sesuai."
                    "\n - Stimulasi mental: Terus melibatkan otak dalam aktivitas yang menantang secara kognitif dapat membantu mempertahankan fungsi kognitif yang ada."
                    "\n - Dukungan sosial: Dukungan dari keluarga, teman, atau kelompok dukungan dapat membantu penderita dan keluarga menghadapi tantangan terkait penyakit ini."
                    "\n - Pendidikan dan penyuluhan: Memberikan informasi tentang penyakit ini kepada penderita dan keluarga dapat membantu mengatasi ketidakpastian dan memberikan panduan dalam merencanakan perawatan dan dukungan")




page_names_to_funcs = {
    "Headline": main_page,
    "Prediksi Tanda Penyakit Alzheimer": alzheimer,
}

selected_page = st.sidebar.selectbox("Select a page", page_names_to_funcs.keys())
page_names_to_funcs[selected_page]()
```
